```python
from bs4 import BeautifulSoup
from urllib import request
```

```python
url = 'https://www.aozora.gr.jp/cards/000148/files/2371_13943.html'
response = request.urlopen(url)
soup = BeautifulSoup(response)
response.close()

print(soup)
```

```python
#本文のテキストだけ取得
main_text = soup.find('div', class_='main_text')
print(main_text)
```

```python
tags_to_delete = main_text.find_all(['rp','rt'])
for tag in tags_to_delete:
  tag.decompose()

print(main_text)
```

```python
import re
main_text = main_text.get_text()
main_text = re.sub(r"[\u3000 \n \r]", "", main_text)
print(main_text)
```

```python
import MeCab
mecab_tagger = MeCab.Tagger()
text = "近頃は大分方々の雑誌から談話をしろしろと責められて、頭ががらん胴になったから、当分品切れの看板でも懸けたいくらいに思っています。"

text_mecab = mecab_tagger.parse(text)
print(text_mecab)
```

```python
text_split = ['近頃','は','大分','方々','の','雑誌','から','談話','を','しろ','しろ','と','責め','られ','て','、','頭','が','がらん','胴','に','なっ','た','から','、','当分','品切れ','の','看板','で','も','懸け','たい','くらい','に','思っ','て','い','ます','。']
```

```python
#ストップワードの除去
url =  'http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt'
response = request.urlopen(url)
soup = BeautifulSoup(response)
response.close()
```

```python
stopwords_text = soup.text
print(stopwords_text)
```

```python
stopwords_list = stopwords_text.split("\r\n")
stopwords_list = [word for word in stopwords_list if word]
print(stopwords_list)
```

```python
result_text_list = list()
for split_text in text_split:
  if split_text not in stopwords_list:
    result_text_list.append(split_text)

print(result_text_list)
```
